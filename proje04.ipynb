```python
from google.colab import drive
drive.mount('/content/drive')


dosyaGidis = "/content/drive/MyDrive/Colab Notebooks/veritabani_txt.ipynb adlı not defterinin kopyası"

def kayıtlarListeleme():
  dosya = open(dosyaGidis, "r")
  for line in dosya.readlines():
    print(line)
    dosya.close()

def yeniKayitEkle():
  kitapAdı = input("Kitap adı giriniz: ")
  yazarAdı = input("Yazar adı giriniz: ")
  basımYılı = input("Basım yılını giriniz: ")
  yayınEvı = input("Yayınevini giriniz: ")
  dosya = open(dosyaGidis, "a") #mevcut veriye ek veri yazılması için parametre "a"
  dosya.write(kitapAdı + " - "+ yazarAdı + " - " + basımYılı + " - " + yayınEvı + " - " + "\n")  
  dosya.close()
  print("\n" + "Kitap ekleme işlemi gerçekleşmiştir" + "\n")
  menu()

  
def kayitAra():
  dosya = open(dosyaGidis)
  arananKitap = input("Aradığınız kitabın adını giriniz: ")
  dosya = open(dosyaGidis, "r")
  arananKitaplar = dosya.read().find(arananKitap)
  if arananKitaplar:
    print("Aradığınız kitap mevcut.")
  elif arananKitaplar:
    print("Aradığınız kitap bulunmamaktadır.")


def menu():
  while True:
    print("***Kütüphane Kataloğuna Hoşgeldiniz.***")
    menuSecim = input("\n---------------------- \n Yapmak istediğiniz işlemi giriniz. \n 1)Kayıtları Listele \n 2)Yeni kayıt ekle \n 3)Kayıt ara \n 4)Çıkış \n")
    
    if menuSecim == "1":
      print("Kayıtları Listele")
      kayıtlarListeleme()
    
    elif menuSecim == "2":
      print("Yeni kayıt ekle")
      yeniKayitEkle()

    elif menuSecim == "3":
      print("Kayıt Ara")
      kayitAra()

    elif menuSecim == "4":
      print("Çıkış Yapıldı")
      break
    else:
      ("\n---------------------- \n Yapmak istediğiniz işlemi giriniz. \n 1)Kayıtları Listele \n 2)Yeni kayıt ekle \n 3)Kayıt ara \n 4)Çıkış \n")
          

menu()




    
```
